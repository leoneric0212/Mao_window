```python
import requests
from requests import Response
aqi_url="https://data.moenv.gov.tw/api/v2/aqx_p_488?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=datacreationdate desc&format=JSON"
try:
    data:Response=requests.request('get',aqi_url)   #requests.request回傳的資料格式為Rseponse
except Exception as e:
    print()
if data.ok:             #回傳布林值
    print("下載成功")
        #執行下載後的資料
else:
    print("下載失敗")
```

```python
print(data.text)
```

```python
try:            #用於確保不因格式錯誤導致的crash
    json_data:dict=data.json()        #json文件本質上是dict
except requests.exceptions.JSONDecodeError:
    print("json格式不正確")

sites:list=json_data['records']   #json文件中資料存在records內
for s in sites:
    print(s)
```

```python
for s in sites:
    print(s['sitename'],s['county'],s['aqi'],s['status'])
    print(type(s['aqi']))       #書出資料為數字，但是型態是str
```
