```python
import matplotlib.pyplot as plt
import numpy as np
def func(pct,allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return f"{absolute:d}pcs - {pct:.1f}%"

value=[5,6]
labels=['rent','return']
colors=['green','red']
figure=plt.figure(figsize=(5,5),dpi=72)
axes=figure.add_subplot()
#製作圓餅圖
axes.pie(value,labels=labels,           
         colors=colors,
         shadow=True,
         startangle=180,
         labeldistance=0.4,
         autopct=lambda pct:func(pct,value),
         textprops=dict(color="b"))         #設顏色
axes.set_title("Ttttitle")
'''
axes.legend(title="ingredients",            #標籤說明
            loc="center left",
            bbox_to_anchor=(1,0,0.5,1))
'''
plt.show()
```
